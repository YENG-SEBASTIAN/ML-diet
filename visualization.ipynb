```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('food_diet.csv')

# List of numerical features
numerical_features = ['Calories (kcal)', 'Protein (g)', 'Carbs (g)', 'Fat (g)', 'BMI']

# Histograms for each numerical feature
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[feature], bins=30, kde=True, color='skyblue')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('histograms.png')
plt.show()

# Scatter plot for Calories vs. Protein with color coding based on Meal Type
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Calories (kcal)'], y=data['Protein (g)'], hue=data['Meal Type'], palette='Set2', edgecolor='w')
plt.title('Scatter Plot: Calories vs. Protein by Meal Type')
plt.xlabel('Calories (kcal)')
plt.ylabel('Protein (g)')
plt.legend(title='Meal Type')
plt.savefig('scatter_calories_protein_meal_type.png')
plt.show()

# Scatter plot for Carbs vs. Fat with color coding based on Meal Type
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Carbs (g)'], y=data['Fat (g)'], hue=data['Meal Type'], palette='Set2', edgecolor='w')
plt.title('Scatter Plot: Carbs vs. Fat by Meal Type')
plt.xlabel('Carbs (g)')
plt.ylabel('Fat (g)')
plt.legend(title='Meal Type')
plt.savefig('scatter_carbs_fat_meal_type.png')
plt.show()

# Scatter plot for Calories vs. Protein with color coding based on BMI Category
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Calories (kcal)'], y=data['Protein (g)'], hue=data['BMI Category'], palette='coolwarm', edgecolor='w')
plt.title('Scatter Plot: Calories vs. Protein by BMI Category')
plt.xlabel('Calories (kcal)')
plt.ylabel('Protein (g)')
plt.legend(title='BMI Category')
plt.savefig('scatter_calories_protein_bmi_category.png')
plt.show()

# Scatter plot for Carbs vs. Fat with color coding based on BMI Category
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Carbs (g)'], y=data['Fat (g)'], hue=data['BMI Category'], palette='coolwarm', edgecolor='w')
plt.title('Scatter Plot: Carbs vs. Fat by BMI Category')
plt.xlabel('Carbs (g)')
plt.ylabel('Fat (g)')
plt.legend(title='BMI Category')
plt.savefig('scatter_carbs_fat_bmi_category.png')
plt.show()

# Box plots for numerical features
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x=data[feature], color='lightgreen')
    plt.title(f'Box Plot of {feature}')
    plt.xlabel(feature)
plt.tight_layout()
plt.savefig('boxplots.png')
plt.show()

# Correlation heatmap of numerical features
plt.figure(figsize=(8, 6))
correlation = data[numerical_features].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.savefig('correlation_heatmap.png')
plt.show()
```
